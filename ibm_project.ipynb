## Importing Libraries 

```python
import numpy as np
import pandas as pd
```

## Loading Data

```python
data1 = pd.read_csv("mental-and-substance-use-as-share-of-disease -AI.csv")
data2 = pd.read_csv("prevalence-by-mental-and-substance-use-disorder _AI.csv")
```

```python
data1
```

```python
data2
```

## Data Preprocessing & Cleaning

```python
data = pd.merge(data1, data2)
data
```

```python
data.isnull().sum()
```

```python
data.drop('Code',axis=1, inplace=True)
data.drop('Year',axis=1, inplace=True)
data
```

```python
# Pop the column and store it in a variable
column_to_move = data.pop('DALYs (Disability-Adjusted Life Years) - Mental disorders - Sex: Both - Age: All Ages (Percent)')

# Assign the popped column back to the DataFrame
data['DALYs (Disability-Adjusted Life Years) - Mental disorders - Sex: Both - Age: All Ages (Percent)'] = column_to_move
```

```python
data
```

```python
#data = data.set_axis(['Country', 'Year', 'Schizophrenia', 'Bipolar', 'Eating', 'Anxiety', 'Drug', 'Depressive', 'Alcohol', 'mental_fitness'], axis=1, copy=False)
data = data.set_axis(['Country', 'Schizophrenia', 'Bipolar', 'Eating', 'Anxiety', 'Drug', 'Depressive', 'Alcohol', 'mental_fitness'], axis=1, copy=False)
```

```python
data.info()
```

## Changing all labels to Numric Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l = LabelEncoder()
for i in data.columns:
    if data[i].dtype == 'object':
        data[i] = l.fit_transform(data[i])
```

```python
data
```

## Splitting Data into Train & Dev Sets

```python
shuffled_data = data.sample(frac=1, random_state=41)
```

```python
shuffled_data
```

```python
X_train = np.array(shuffled_data.iloc[:5472, 0:8])
X_dev = np.array(shuffled_data.iloc[5472:6156, 0:8])
X_test = np.array(shuffled_data.iloc[6156:, 0:8])
print("X_train : ",X_train.shape)
print("X_dev : ",X_dev.shape)
print("X_test : ",X_test.shape)
```

```python
Y_train = np.array(shuffled_data.iloc[:5472, 8]).reshape(-1,1)
Y_dev = np.array(shuffled_data.iloc[5472:6156, 8]).reshape(-1,1)
Y_test = np.array(shuffled_data.iloc[6156:, 8]).reshape(-1,1)
print("Y_train : ",Y_train.shape)
print("Y_dev : ",Y_dev.shape)
print("Y_test : ",Y_test.shape)
```

## Model 

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

### Architecture

```python
model = Sequential([
    Dense(units=150, activation='sigmoid', input_shape=(8,)),
    Dense(units=1, activation='linear')
])
```

```python
model.summary()
```

### Compiling Model

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

### Train Model

```python
history = model.fit(x=X_train, y=Y_train, validation_data=(X_dev, Y_dev), epochs=40)
```

```python
import matplotlib.pyplot as plt
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'b', label='Training Loss')
plt.plot(epochs, val_loss, 'r', label='Validation Loss')
plt.title('Epochs vs. Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

## Evaluation

```python
# Evaluate model performance on test dataset
loss = model.evaluate(X_test, Y_test)

# Print the evaluation results
print("Test Loss:", loss)
```

```python
# Print the predicted and actual values side by side
y_pred = model.predict(X_test)

for i in range(len(X_test)):
    print("Predicted:", y_pred[i], "\tActual:", Y_test[i])
```
